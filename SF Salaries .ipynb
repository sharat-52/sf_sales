```python
# Replace the file path with your actual file path
file_path = "E:/Professional/AI/Python practice/SF Salaries/Salaries.csv"
salaries_df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify the import
salaries_df.head()
```

```python
# List of columns to be removed
columns_to_remove = ["Notes", "Agency", "Status"]

# Drop the specified columns
salaries_df = salaries_df.drop(columns=columns_to_remove)

# Verify that the columns have been removed
print(salaries_df.head())
```

```python
statistics = salaries_df.describe(include='all')
print(statistics)
```

```python
# Count the occurrence of each employee name
employee_name_counts = salaries_df['EmployeeName'].value_counts()

# Get the top 5 most common names
top_5_names = employee_name_counts.head(5)

print(top_5_names)
```

```python
unique_job_titles = salaries_df['JobTitle'].nunique()
print("Number of unique job titles:", unique_job_titles)
```

```python
# Filter the DataFrame to include only rows where 'JobTitle' contains 'captain' (case-insensitive)
captain_jobs = salaries_df[salaries_df['JobTitle'].str.contains('captain', case=False)]

# Get the total number of job titles containing 'captain'
total_captain_jobs = captain_jobs.shape[0]
print("Total number of job titles containing 'captain':", total_captain_jobs)
```

```python
import numpy as np  # Import numpy for handling NaN values

# Replace 'Not Provided' with NaN in the 'EmployeeName' column
salaries_df['EmployeeName'] = salaries_df['EmployeeName'].replace('Not Provided', np.nan)

# Verify the replacement
print(salaries_df.head())
```

```python
# Count the occurrence of each job title
job_title_counts = salaries_df['JobTitle'].value_counts()

# Get the top 5 most common job titles
top_5_job_titles = job_title_counts.head(5)

print(top_5_job_titles)
```

```python
import matplotlib.pyplot as plt

# Filter the dataset for the "Transit Operator" job title
transit_operator_df = salaries_df[salaries_df['JobTitle'] == 'Transit Operator']

# Group the data by year and calculate the sum of total pay for each year
yearly_totalpay = transit_operator_df.groupby('Year')['TotalPay'].sum()

# Create a time series plot
plt.figure(figsize=(10, 6))
plt.plot(yearly_totalpay.index, yearly_totalpay.values, marker='o', linestyle='-')
plt.title('Total Pay for "Transit Operator" Over Time')
plt.xlabel('Year')
plt.ylabel('Total Pay')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter the dataset for 'Transit Operator' job title
transit_operator_df = salaries_df[salaries_df['JobTitle'] == 'Transit Operator']

# Group by 'Year' and calculate the average 'TotalPay'
average_totalpay_by_year = transit_operator_df.groupby('Year')['TotalPay'].mean()

# Create a time series plot
plt.figure(figsize=(10, 6))
plt.plot(average_totalpay_by_year.index, average_totalpay_by_year.values, marker='o')
plt.title('Average TotalPay for Transit Operator Over the Years')
plt.xlabel('Year')
plt.ylabel('Average TotalPay')
plt.grid(True)
plt.show()
```
